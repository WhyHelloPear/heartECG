## Data description

There are  87553 observations in the training set and 21891 observations in the test set.  All data is scaled between 0 and 1.  Each observation has 187 features.  The features are correspond to a human heart ECG signal sampled at 125hz.  Thus the data is time dependent.  The labels are included in the last column.  Each label can take on a value between 0 and 4.  Zero meaning normal and 1 - 4 meaning abnormal.  More to be added about abnormal conditions.  We would like to classify these time dependent signals into each of the 5 classes.
